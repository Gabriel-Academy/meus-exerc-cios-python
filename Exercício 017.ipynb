```python
from math import hypot

n = float(input('Digite o número do cateto oposto: '))
n2 = float(input('Digite o número do cateto adjacente: '))
n3 = hypot(n, n2)
print(f'O valor da hipotenusa é igual a {n3:.2f}')
```
