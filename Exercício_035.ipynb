```python
print('-='*20)
print('Avaliador de Triângulos')
print('-='*20)
n = float(input('Digite o primeiro comprimento: '))
n2 = float(input('Digite o segundo comprimento: '))
n3 = float(input('Dgite o terceiro comprimento: '))
if n < n2 + n3 and n2 < n + n3 and n3 < n + n2:
  print('Os comprimento acima PODEM formar um triângulo')
else:
  print('Os comprimentos acima NÃO podem formar um triângulo')
```
