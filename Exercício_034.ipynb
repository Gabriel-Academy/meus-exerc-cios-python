```python
n = float(input('Digite o valor do seu salário: '))
print(f'Show, vamos calcular as porcentagens do seu salário de R${n}')
if n <=1250:
  n2 = (n*15) / 100
  print(f'Parabéns pelo aumento, você está apto a receber aumento de 15%! Seu salário atual passará a ser R${n2+n}')
else:
  n2 = (n*10) / 100
  print(f'Parabés pelo aumento, você está apto a receber o aumento de 10%! Seu salário atual passará a ser R${n2+n}')
```
