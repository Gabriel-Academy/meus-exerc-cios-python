```python
a = int(input('Digite um número: '))
b = int(input('Digite o segundo número: '))
c = int(input('Digite o terceiro número: '))

#Menor

menor = a
if b < a and b < c:
  menor = b
if c < a and c < b:
  menor = c

# Maior

maior = a
if b > a and b > c:
  maior = b
if c > a and c > b:
  maior = c

print(f'Dentre os números digitados, é notório que {menor} é o menor número.')

print(f'Dentre os números digitados, é notótio que {maior} é o maior número. ')
```
