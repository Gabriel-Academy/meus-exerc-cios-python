```python

n = float(input('Digite a velocidade atual do seu carro: '))
multa = (n-80) * 7
if n <=80:
  print(f'Perfeito, você está dentro da velocidade ideal! 80Km/h')
else:
  print(f'Você ultrapassou o limiite de velocidade 80Km/h e foi multado com R${multa:.2f}. Tenha um ótimo dia, dirija com segurança.')
```
