```python
from math import sin, radians, cos, tan

n = float(input('Digite o ângulo: '))

n1 = sin(radians(n))

print(f'O valor de SENO do ângulo {n:.0f}° é igual a {n1:.2f}')

n2 = cos(radians(n))

print(f'O valor de COSSENO do ângulo {n:.0f}° é igual a {n2:.2f}')

n3 = tan(radians(n))

print(f'O valor da TANGENTE do ângulo {n:.0f}° é igual a {n3:.2f}')
```
